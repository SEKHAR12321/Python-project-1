```python
def add_employee(emp_id,emp_name,emp_DOJ,emp_designation,emp_salary=0):
    file=open("emp.txt",'a')
    l=[emp_id,emp_name,emp_DOJ,emp_designation,emp_salary]
    for i in l:
        file.write(str(i)+'\n')
    file.close()
def add_hr(emp_id,hr_dept,hr_role):
    file=open("hr.txt",'a')
    l=[emp_id,hr_dept,hr_role]
    for i in l:
        file.write(str(i)+'\n')
    file.close()
def search_employee(emp_id):
    file=open("emp.txt",'r')
    flag=0
    empdet=[i.strip() for i in file.readlines()]
    for i in range(len(empdet)):
        if emp_id==empdet[i]:
            flag=1
            print("\nEmployee details-")
            print("Employee ID",empdet[i])
            print("Employee Name",empdet[i+1])
            print("Employee Date of joining-",empdet[i+2])
            print("Employee designation-",empdet[i+3])
            print("Employee salary-",empdet[i+4])
            break
    if flag==0:
        print("No such Employee id")
    file.close()
def search_hr(emp_id):
    file=open("hr.txt",'r')
    flag=0
    hrdet=[i.strip() for i in file.readlines()]
    for i in range(len(hrdet)):
        if emp_id==hrdet[i]:
            flag=1
            print("\nHR details-")
            print("Employee ID",hrdet[i])
            print("Hr Department-",hrdet[i+1])
            print("Hr Role-",hrdet[i+2])
            break
    if flag==0:
        print("No such employee id")
    file.close()
def upd_employee(emp_id):
    file=open("emp.txt",'r')
    empdet=file.readlines()
    flag=0
    for i in range(len(empdet)):
        if emp_id==empdet[i].strip():
            flag=1
            print("\nEmployee Current Details-")
            print("Employee Id-",empdet[i])
            print("Employee Designation-",empdet[i+3])
            print("Employee salary",empdet[i+4])
            print("\nEnter the new values-")
            n1=input("Enter Employee new designation-")
            n2=input("Enter Employee new salary-")
            empdet[i+3]=str(n1)+'\n'
            empdet[i+4]=str(n2)+'\n'
            file.close()
            file=open("emp.txt",'w')
            file.writelines(empdet)
            file.close()
            break
    if flag==0:
        print("No such employee id")

        
print("........EMPLOYEE DATABASE.........")
n=1
while(n!=6):
    print("\n1.ADD EMPLOYEE")
    print("2.SEARCH EMPLOYEE")
    print("3.ADD HR")
    print("4.SEARCH HR")
    print("5.UPDATE EMPLOYEE")
    print("6.EXIT")
    n=int(input("\nEnter your choice"))
    if n==1:
        a=input("Enter Employee ID-")
        b=input("Enter Employee Name-")
        c=input("Enter Employee date of joining-")
        d=input("Enter Employee designation-")
        e=input("Enter Employee salary-")
        add_employee(a,b,c,d,e)
    elif n==2:
        a=input("Enter Employee id to search-")
        search_employee(a)
    elif n==3:
        a=input("Enter Employee ID-")
        b=input("Enter HR department-")
        c=input("Enter Hr role")
        add_hr(a,b,c)
    elif n==4:
        a=input("Enter the employee id to search-")
        search_hr(a)
    elif n==5:
        a=input("enter the empoyee id to update-")
        upd_employee(a)
    
```
